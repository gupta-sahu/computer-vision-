```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('Ganeshji.jpg')
```

```python
xyz=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
# Apply Adaptive Gaussian Thresholding
gaussian_thresh = cv2.adaptiveThreshold(
    xyz, 255, 
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
    cv2.THRESH_BINARY, 
    11, 2
)
```

```python
# Visualization
plt.figure(figsize=(10, 5))

# Original grayscale image
plt.subplot(1, 2, 1)
plt.imshow(xyz, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis('off')
```

```python
# Adaptive Gaussian Thresholded Image
plt.subplot(1, 2, 2)
plt.imshow(gaussian_thresh, cmap='gray')
plt.title("Adaptive Gaussian Thresholding")
plt.axis('off')
```

```python
plt.tight_layout()
plt.show()
```
