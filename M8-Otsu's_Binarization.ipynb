```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('marksheet.jpeg')
```

```python
xyz=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
blurred_image = cv2.GaussianBlur(xyz, (5,5), 0)
```

```python
abc = cv2.adaptiveThreshold(xyz,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,blockSize=11,C=3)
```

```python
_, otsu_thresh = cv2.threshold(xyz, 0, 255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
TH=100
max_TH=255
_, Threshold=cv2.threshold(blurred_image,TH,max_TH,cv2.THRESH_BINARY)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,5,1)
plt.title("Original Image:")
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,5,2)
plt.title("B&W Image")
plt.imshow(xyz, cmap='gray')
plt.axis('off')

plt.subplot(1, 5, 3)
plt.title("Binary Image (Thresholding)")
plt.imshow(Threshold / 255.0, cmap='gray')
plt.axis('off')

plt.subplot(1,5,4)
plt.title("Adaptive")
plt.imshow(abc,cmap="gray")
plt.axis("off")

plt.subplot(1, 5, 5)
plt.title("Otsu's Binarization")
plt.imshow(otsu_thresh, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
