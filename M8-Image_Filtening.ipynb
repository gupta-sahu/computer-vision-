```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Read the image
image=cv2.imread('flower.jpg')
```

```python
# Gaussian Blue

#Apply Gaussian Blur
#(5,5) is the size of the kernel, 0 let OpenCV pick sigma automatically
blurred = cv2.GaussianBlur(image,(5,5),0)
```

```python
# Median Blur 

median_blurred = cv2.medianBlur(image,5)
```

```python
# Bilateral Filtering
#Argumkent: (image, dimester, sigmaColour, sigmaSpace)

filtered =cv2.bilateralFilter(image, d=9, sigmaColor=75, sigmaSpace=75)
```

```python
# Sharpening Filters

# create the mask(difference)
mask = cv2.subtract(image, blurred)

# Add thr mask back to the original image(sharpen)
shaepened = cv2.addWeighted(image, 1.0, mask, 1.5,0)
```

```python
plt.subplot(1,5,1)
plt.title("Original Image:")
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1,5,2)
plt.title("Gaussian Blurred")
plt.imshow(cv2.cvtColor(blurred,cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1,5,3)
plt.title("Median Blurred")
plt.imshow(cv2.cvtColor(median_blurred,cv2.COLOR_BGR2RGB))
plt.axis('off')


plt.subplot(1,5,4)
plt.title("Bilateral Filtering")
plt.imshow(cv2.cvtColor(filtered,cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1,5,5)
plt.title("Shaepened")
plt.imshow(cv2.cvtColor(shaepened,cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```
